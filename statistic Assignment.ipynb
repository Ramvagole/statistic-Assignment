```python
#Q3):-
type of data are
1)Quantitative:-
  Quantitative data can be expressed in numerical values, making it countable and including statistical data analysis.
  These kinds of data are also known as Numerical data. It answers the questions like “how much,” “how many,” and “how often.” For example, the price of a phone, the computer’s ram, the height or weight of a person, etc., falls under quantitative data. 
  Quantitative data can be used for statistical manipulation.
  These data can be represented on a wide variety of graphs and charts, such as bar graphs, histograms, scatter plots, boxplots, pie charts, line graphs, etc.
  
    Discrete Data:-
    The term discrete means distinct or separate. 
    The discrete data contain the values that fall under integers or whole numbers. 
    The total number of students in a class is an example of discrete data. 
    These data can’t be broken into decimal or fraction values.
    
    Continuous Data:-
    Continuous data are in the form of fractional numbers. 
    It can be the version of an android phone, the height of a person, the length of an object, etc. 
    Continuous data represents information that can be divided into smaller levels. 
    The continuous variable can take any value within a range. 
    
2)Qualitative or Categorical Data:-
  Qualitative or Categorical Data is data that can’t be measured or counted in the form of numbers. 
  These types of data are sorted by category, not by number. 
  That’s why it is also known as Categorical Data. 
  These data consist of audio, images, symbols, or text.
  The gender of a person, i.e., male, female, or others, is qualitative data.
  Qualitative data tells about the perception of people. 
  This data helps market researchers understand the customers’ tastes and then design their ideas and strategies accordingly. 
    
   Nominal Data:-
   Nominal Data is used to label variables without any order or quantitative value. 
   The color of hair can be considered nominal data, as one color can’t be compared with another color.
  
   Ordinal Data:-
   Ordinal data have natural ordering where a number is present in some kind of order by their position on the scale. 
   These data are used for observation like customer satisfaction, happiness, etc., but we can’t do any arithmetical tasks on them. 
```

```python
#Q5):-
Nominal level:-
  You can categorize your data by labelling them in mutually exclusive groups, but there is no order between the categories.
   Example:-
    City of birth
    Gender
    Ethnicity
    Car brands
    Marital status
    
Ordinal level:-
 You can categorize and rank your data in an order, but you cannot say anything about the intervals between the rankings.
 Although you can rank the top 5 Olympic medallists, this scale does not tell you how close or far apart they are in number of wins.
 Example:-
  Top 5 Olympic medallists
  Language ability (e.g., beginner, intermediate, fluent)
  Likert-type questions  (e.g., very dissatisfied to very satisfied)

Interval level:-
  You can categorize, rank, and infer equal intervals between neighboring data points, but there is no true zero point.
  The difference between any two adjacent temperatures is the same: one degree. But  zero degrees is defined differently depending on the scale – it doesn’t mean an absolute absence of temperature.
  The same is true for test scores and personality inventories. A zero on a test is arbitrary; it does not mean that the test-taker has an absolute lack of the trait being measured.
  Example:-
   Test scores (e.g., IQ or exams)
   Personality inventories
   Temperature in Fahrenheit or Celsius
    
Ratio level:-
 You can categorize, rank, and infer equal intervals between neighboring data points, and there is a true zero point.
 A true zero means there is an absence of the variable of interest. In ratio scales, zero does mean an absolute lack of the variable.
 For example, in the Kelvin temperature scale, there are no negative degrees of temperature – zero means an absolute lack of thermal energy.
 Example:-
  Height
  Age
  Weight
  Temperature in Kelvin
```

```python
#Q6):-
Understanding the level of measurement is crucial when analyzing data for several reasons:

1):Appropriate statistical analysis: Different levels of measurement require different statistical techniques. 
   Choosing the wrong analysis can lead to incorrect results and misleading conclusions.
   By understanding the level of measurement, researchers can apply the appropriate statistical tests that are best suited for the data at hand. 
   This ensures that the analysis is valid and provides accurate insights.

2):Interpretation of results: The level of measurement affects the type of conclusions that can be drawn from the data. 
   Certain statistical measures and tests are only meaningful and interpretable at specific levels of measurement. 
   Understanding the level of measurement helps researchers correctly interpret the results and draw appropriate conclusions based on the properties of the data.

Here's an example to illustrate the importance of understanding the level of measurement:

 Suppose a researcher is studying customer satisfaction in a retail store using a survey. 
 One question in the survey asks customers to rate their satisfaction on a scale of 1 to 5, with 1 being "very dissatisfied" and 5 being "very satisfied."
 The researcher collects responses from a sample of customers.

 If the researcher treats the satisfaction ratings as interval-level data and calculates the mean satisfaction score across the sample, it implies that the difference between a rating of 4 and 5 is the same as the difference between 2 and 3. 
 However, in reality, the intervals between the categories may not be equal, and a rating of 4 may not represent twice the satisfaction of a rating of 2. 
 Treating the ratings as interval-level data and interpreting the mean without considering the level of measurement would lead to misleading conclusions.
 In this case, the satisfaction ratings are actually at the ordinal level because the intervals between the categories may not be equally spaced. 
 To analyze the data appropriately, the researcher should use non-parametric tests like the Mann-Whitney U test or rank correlation measures instead of relying solely on the mean. 
 Understanding the level of measurement helps the researcher choose the correct statistical analysis and draw accurate conclusions about customer satisfaction in the retail store. 
```

```python
#Q7):-
Nominal Data:

Nominal data is the lowest level of measurement.
It represents categories or labels that have no inherent order or numerical significance.
The categories are mutually exclusive and exhaustive, meaning each observation can be assigned to only one category, and all possible categories are covered.
Examples of nominal data include gender (male/female), eye color (blue/brown/green), or types of vehicles (car/truck/motorcycle).

Ordinal Data:

Ordinal data is the second level of measurement, above nominal data.
It represents categories that can be ranked or ordered.
The categories have a relative order, indicating that one category is greater or lesser than another.
However, the intervals between the categories may not be equal or quantifiable.
Examples of ordinal data include ratings or rankings (e.g., Likert scales), educational levels (high school diploma/bachelor's degree/master's degree), or customer satisfaction levels (very dissatisfied/dissatisfied/neutral/satisfied/very satisfied).
```

```python
#Q9):-
Descriptive Statistics:

Descriptive statistics involves summarizing and describing the main features of a dataset. It focuses on organizing, displaying, and summarizing the data in a meaningful and concise way.
Descriptive statistics aims to provide a clear and accurate understanding of the data, without making any generalizations or drawing conclusions beyond the specific dataset being analyzed.
Common measures used in descriptive statistics include measures of central tendency (e.g., mean, median, mode), measures of dispersion (e.g., range, standard deviation), and graphical representations (e.g., histograms, bar charts).
Example: Suppose you have a dataset consisting of the ages of 100 individuals in a population. Descriptive statistics can be used to calculate the mean age, median age, and standard deviation to summarize the central tendency and variability of the ages in the sample. Additionally, a histogram or bar chart can be created to visually display the age distribution.

Inferential Statistics:

Inferential statistics involves making inferences, predictions, or generalizations about a population based on sample data. It aims to draw conclusions beyond the specific dataset by using probability theory and statistical inference techniques.
Inferential statistics uses sample data to make inferences about the population parameters and test hypotheses or relationships between variables.
By analyzing the sample data, inferential statistics allows researchers to make generalizations and draw conclusions about the larger population from which the sample was drawn.
Example: Suppose a pharmaceutical company wants to test the effectiveness of a new drug compared to a placebo. They randomly assign participants to two groups: one receiving the new drug and the other receiving a placebo. After the treatment period, inferential statistics can be used to compare the mean outcomes (e.g., reduction in symptoms) between the two groups and determine if the observed difference is statistically significant. This analysis allows the researchers to infer whether the new drug is truly effective compared to the placebo for the larger population of interest.
```

```python
#Q10):-
Measures of Central Tendency:

Mean: The mean is the arithmetic average of a dataset. It is calculated by summing all the values and dividing by the total number of observations. The mean provides a measure of the typical or average value in the dataset. It is influenced by extreme values and is appropriate for datasets with a symmetrical distribution.

Median: The median represents the middle value of a dataset when it is ordered from smallest to largest. If there is an even number of observations, the median is the average of the two middle values. The median is less affected by extreme values and is suitable for datasets with skewed distributions or outliers.

Mode: The mode is the most frequently occurring value in a dataset. It represents the value(s) that appear(s) with the highest frequency. The mode is useful for identifying the most common category or value in categorical data.

Measures of Variability:

Range: The range is the difference between the maximum and minimum values in a dataset. It provides a simple measure of the spread or variability in the data. However, it is sensitive to extreme values and may not fully capture the distribution's shape.

Variance: Variance measures the average squared deviation of each data point from the mean. It provides an indication of how spread out the data is. A higher variance suggests greater variability, while a lower variance indicates less dispersion.

Standard Deviation: The standard deviation is the square root of the variance. It measures the average distance between each data point and the mean, providing a standardized measure of the dataset's spread. Like variance, a higher standard deviation implies greater variability.
```
